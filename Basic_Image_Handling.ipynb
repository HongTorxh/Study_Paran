```python
import numpy as np
import os
from imageio import imread
from scipy.misc import imresize
import matplotlib.pyplot as plt
%matplotlib inline
print("PACKAGE LOADED.")
```

```python
cwd = os.getcwd()
print("Current folder is %s" % (cwd))
```

```python
def print_typeshape(img):
  print("Type is %s" % (type(img)))
  print("Shape is %s" % (img.shape,))
```

```python
cat = imread("./data/cat.jpeg")
print_typeshape(cat)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(cat)
plt.title("ORIGINAL CAT")
plt.draw()
```

```python
cat2 = imread("./data/cat.jpeg").astype(np.float)
print_typeshape(cat2)
```

```python
plt.figure(figsize=(10,8))
plt.imshow(cat2)
plt.title("Original Image with imread.astype(np.float)")
plt.draw()
```

```python
cat3 = imread("./data/cat.jpeg").astype(np.float)/255.
print_typeshape(cat2)

plt.figure(figsize=(10,8))
plt.imshow(cat3)
plt.title("Original Image with imread.astype(np.float)")
plt.draw()
```

```python
catsmall = imresize(cat, [100,100])
print_typeshape(catsmall)

plt.figure(1)
plt.imshow(catsmall)
plt.title("Resize Image")
plt.draw()
```

```python
def rgb2gray(rgb):
  if len(rgb.shape) is 3:
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])
  else:
    print("Current Image if GRAY!")
    return rgb
```

```python
catsmallgray = rgb2gray(catsmall)
print("size of catsmallgray is %s" % (catsmallgray.shape,))
print("type of catsmallgray is", type(catsmallgray))
```

```python
plt.imshow(catsmallgray) #cmap이 없을 때
# why ? 행렬을 그리고 있기 때문인데, clolormap이라는 것을 통해 각각의 값을 색깔로 표현해줘야됨.
plt.title("[imshow] Gray Image")
plt.colorbar()
plt.draw
```

```python
plt.imshow(catsmallgray, cmap=plt.get_cmap("gray")) #camp이 있을 때
plt.title("[imshow] Gray Image")
plt.colorbar()
plt.draw
```

```python
# Reshape : 모양을 바꾸는 것, 모양을 바꿀 때 element의 갯수는 변하면 안된다.
# (1, -1) : 한줄로 피되, 그 옆에 있는 것은 알아서 펴라
catrowvec = np.reshape(catsmallgray, (1,-1));
print("size of catrowvec is %s" % (catrowvec.shape,))
print("type of catrowvec is", type(catrowvec))


catmatrix = np.reshape(catrowvec, (100,-1));
print("size of catmatrix is %s" % (catmatrix.shape,))
print("type of catmatrix is", type(catmatrix))
```

```python
plt.imshow(catmatrix, cmap=plt.get_cmap("gray")) #camp이 있을 때
plt.title("Reshaped Matrix")
plt.colorbar()
plt.draw
```

```python
cwd = os.getcwd()
path = cwd + "/img/"
flist = os.listdir(path)
print("[%d] FILE ARE IN [%s]" % (len(flist), path))
```

```python
flist
```

```python
for i, f in enumerate(flist):
  print("[%d]TH FILE IS [%s]" % (i, f))
```

```python
valid_exts = [".jpg", ".gif", ".png", ".tga", ".jpeg"]
imgs = []
names = []

for f in flist:
  ext = os.path.splitext(f)[1]
  if ext.lower() not in valid_exts:
    continue
  fullpath = os.path.join(path, f)
  imgs.append(imread(fullpath))
  names.append(os.path.splitext(f)[0])
```

```python
names
```

```python
for img, name in zip(imgs, names):
  plt.imshow(img)
  plt.title(name)
  plt.show()
```
